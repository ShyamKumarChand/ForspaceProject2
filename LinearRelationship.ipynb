```python
x = np.arange(-5.0, 5.0, 0.1)

## You can adjust the slope and intercept to  verify the changes in the graph
y = 1*(x**3)  + 1*(x**2) + 1*x + 3
y_noise = 5 * np.random.normal(size=x.size)
ydata = y + y_noise

# Plotting My X and Y
plt.figure(figsize=(16,9))
plt.plot(x, ydata,   'bo')
plt.plot(x,y,  'k')
plt.title("cubic of Poly")
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)

## You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x, 2)
y_noise = 1 *np.random.normal(size=x.size)
ydata = y + y_noise

plt.figure(figsize=(16,9))
plt.plot(x, ydata,   'bo')
plt.plot(x,y,  'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.title('Quadratic or plob')
plt.show()
```

```python
X = np.arange(-5.0, 5.0, 0.1)


## You can adjust the slope and intercept to verify the changes in the graph

Y = np.exp(X)
y_noise = 5 *np.random.normal(size=X.size)
ydata = Y + y_noise
```

```python
plt.figure(figsize=(16,9))
plt.plot(X, ydata,   'yo')
plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.title('Exponential')
plt.show()
```

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = np.log(X)

y_noise = 0.2 *np.random.normal(size=X.size)
ydata = Y + y_noise

plt.figure(figsize=(16,9))
plt.plot(X, ydata,   'bo')
plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.title('Lograthamic')
plt.show()
```

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = 1-4/(1+np.power(3, X-2))

y_noise = 0.10 *np.random.normal(size=X.size)
ydata = Y + y_noise

plt.figure(figsize=(16,9))
plt.plot(X, ydata,   'bo')
plt.plot(X,Y)

plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.title('Sigmoid')
plt.show()
```
