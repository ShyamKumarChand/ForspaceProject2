```python
x = np.linspace(-1.4,1.4,30)
plt.plot(x, x, 'g--',x, x**2, 'r:',x, x**3, 'b^')
plt.show()
```

```python
x = np.linspace(-1.4,1.4,30)
line1, line2, line3 = plt.plot(x, x, 'g--',x, x**2, 'r:',x, x**3, 'b^')
line1.set_linewidth(9.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

```python
x = np.linspace(-1.4,1.4,30)
plt.plot(x, x**2)
plt.savefig("my_square_function.png",transperant = True)
```

```python
x = np.linspace(-1.4,1.4,30)
plt.subplot(2, 2, 1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x, x**2)
plt.subplot(2,2,3)
plt.plot(x, x**3)
plt.subplot(2,2,4)
plt.plot(x,x**4)
plt.show()
```

```python
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show()
```

```python
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")
plt.figure(1)
plt.plot(x, -x**3, "r:")
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Square function") # Name of the Plot or Title of the plot
plt.xlabel("x-axis")  # X-axis Name
plt.ylabel("y = x**2") # Y-axis Name
plt.grid(True) # Grid Lines in Plot
plt.show() # it Is used to show the Graphs
```
