```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5.0, 5.0, 0.1)
```

```python
x
```

```python
y = 2*(x)+3
y_noise = 0.2*np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
plt.figure(figsize=(16,8))
plt.scatter(x, ydata)
plt.plot(x,y, 'r')
plt.title("Strongly Positive Linear Regresion line")
plt.xlabel("Dependant label")
plt.ylabel("Independant")
plt.show()
```

```python
x = np.linspace(-1, -4, 100)
plt.figure(figsize=(16,8))
plt.scatter(x, ydata)
plt.plot(x,y, 'r')
plt.title("Strongly Positive Linear Regresion line")
plt.xlabel("Dependant label")
plt.ylabel("Independant")
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 5*np.random.normal(size=x.size)
ydata = y + y_noise
plt.figure(figsize=(16,9))
plt.plot(x,y, 'bo')
plt.plot(x,y, 'k')
plt.scatter(x, ydata)
plt.plot(x,y, 'r')
plt.title("Strongly Positive Linear Regresion line")
plt.xlabel("Dependant label")
plt.ylabel("Independant")
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.log(1*(x**3) + 1*(x**2) + 1*x + 3)
y_noise = 0.9*np.random.normal(size=x.size)
ydata = y + y_noise
plt.figure(figsize=(16,9))
plt.plot(x,y, 'bo')
plt.plot(x,y, 'k')
plt.scatter(x, ydata)
plt.plot(x,y, 'r')
plt.title("Strongly Positive Linear Regresion line")
plt.xlabel("Dependant label")
plt.ylabel("Independant")
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.power(x,2)
y_noise = 2*np.random.normal(size=x.size)
ydata = y + y_noise
plt.figure(figsize=(16,9))
plt.plot(x,ydata, 'bo')
plt.plot(x,y, 'r')
plt.title("Quadratic equation")
plt.xlabel("Dependant label")
plt.ylabel("Independant")
plt.show()
```

```python
x = np.linspace(-9, 9, 400)
y = np.linspace(-5, 5, 400)
x,y = np.meshgrid(x,y)
```

```python
def axes():
    plt.figure(figsize=(16,9))
    plt.axhline(0, alpha=.9)
    plt.axvline(0, alpha=.9)
```

```python
a = .4
b = .2
axes()
plt.contour(x, y, (x**2/a**2- y**2/b**2),[1],colors='y')
plt.show()
```

```python
axes()
plt.contour(x, y, (y**2 -4*a*x),[0],colors='y')
```
