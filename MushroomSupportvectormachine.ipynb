```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/reddyprasade/Machine-Learning-Problems-DataSets/master/Classification/mushrooms.csv'
```

```python
Data = pd.read_csv(url)
Data.head()
```

```python
Data_encode = pd.get_dummies(Data)
Data_encode
```

```python
X_mush = Data_encode.iloc[:,2:]
y_mush = Data_encode.iloc[:,1]
```

```python
X_mush.head()
```

```python
y_mush.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_mush
```

```python
y_mush
```

```python
x_train,x_test, y_train, y_test=train_test_split(X_mush,y_mush,test_size=0.30)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear',C=1E10)
```

```python
model.fit(X_mush,y_mush)
```

```python
New_prediction = model.predict(x_test)
New_prediction
```

```python
Comp = pd.DataFrame({'ACTUAL_DATA':y_test,
                    "New_prediction":New_prediction})
```

```python
Comp
```

```python
Train_Score = model.score(x_train,y_train)*100
```

```python
print("Trainning Score is {}".format(Train_Score))
```

```python
Test_Score = model.score(x_test,y_test)*100
print("Testing Score is {}".format(Test_Score))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,New_prediction))
```

```python
cm = confusion_matrix(y_test,New_prediction)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
Kernal = ['linear', 'poly', 'rbf', 'sigmoid']
```

```python
for i in Kernal:
    print("Right Now My Kernal is {}".format(i))
    model_svc = SVC(kernel=i)
    model_svc.fit(x_train,y_train)
    T_Score = model_svc.score(x_train,y_train)
    Ts_Score = model_svc.score(x_test,y_test)
    print("My Train Score for the this kernal is {} \n Test Score for this Kernal is {}".format(T_Score,Ts_Score))
    New_predication = model_svc.predict(x_test)
    Comp = pd.DataFrame({"Actual_Data":y_test,
                 "New_predication":New_predication})
    print(Comp.head(),'\n')
    print("CLASSIFICATION REPORT")
    Report = classification_report(y_test,New_predication)
    print(Report)
    print("CONFUSION MATRIX")
    cm =confusion_matrix(y_test,New_predication)
    sns.heatmap(cm,annot=True)
    plt.show()
```
