```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/reddyprasade/Machine-Learning-Problems-DataSets/master/Classification/mushrooms.csv'
```

```python
Data = pd.read_csv(url)
Data.head()
```

```python
Data_encode = pd.get_dummies(Data)
Data_encode
```

```python
X_mush = Data_encode.iloc[:,2:]
y_mush = Data_encode.iloc[:,1]
```

```python
X_mush.head()
```

```python
y_mush.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import validation_curve
```

```python
def getTrainingandTestScore():
    train_scores, valid_score = validation_curve(SVC(kernel='rbf', C=1, random_state=0), X_mush, y_mush,
                                                param_name="gamma",scoring='accuracy',param_range=np.logspace(-4,1,6),
                                                verbose=3)
    train_scores = np.mean(train_scores, axis=1)
    valid_scores = np.mean(valid_scores, axis=1)
    return(train_scores,valid_scores)
training, test = getTrainingandTestScore()
    
```

```python
print("Training scores:")
print(training)
print("Test scores:")
print(test)
print("on Different Gamma")
print(np.logspace(-4, 1, 6))
```

```python
def getGammaValues():
    gamma = np.logspace(-4,1,6)
    return(gamma[0],gamma[5],gamma[3])
underfit, overfit, good_generalization = getGammaValues()
print("underfitting value")
print(underfit)
print("overfitting value")
print(overfit)
print("good value")
print(good_generalization)
```
